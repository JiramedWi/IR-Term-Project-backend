```python
import pandas as pd
import string
import numpy as np
import csv

lyrics_data = pd.read_csv("./lyrics-data.csv")
artists = pd.read_csv("./artists-data.csv")

lyrics_data.dropna(inplace=True)
artists.dropna(inplace=True)

lyrics_only_en = lyrics_data.drop(
    lyrics_data[lyrics_data['Idiom'] != 'ENGLISH'].index)
lyrics_only_en.shape
```

```python

lyrics_only_en.drop_duplicates(subset='SLink', keep='first', inplace=True, ignore_index=False)
artists = artists.drop_duplicates(subset='Artist')
lyrics_only_en.shape
```

```python
lyrics_nd = lyrics_only_en.drop_duplicates(subset='SName')
lyrics_nd.shape
```

```python
dataAll = pd.merge(left=lyrics_nd, right=artists,
                   left_on='ALink', right_on='Link')
dataAll.tail(10)
```

```python
dataAll = dataAll.drop(columns=['ALink', 'SLink' ,'Idiom', 'Songs', 'Popularity', 'Link', 'Genre', 'Genres'])
dataAll.head(10)
```

```python
lyrics = dataAll['Lyric']
cleanned_lyrics = lyrics.apply(lambda s: s.translate(
    str.maketrans('', '', string.punctuation + u'\xa0')))
cleanned_lyrics = cleanned_lyrics.apply(lambda s: s.lower())
cleanned_lyrics = cleanned_lyrics.apply(lambda s: s.translate(
    str.maketrans(string.whitespace, ' '*len(string.whitespace), '')))

cleanned_lyrics.tail(10)
```

```python
type(cleanned_lyrics)
df = pd.concat([cleanned_lyrics], axis=1)
df
```

```python
dataAll = pd.merge(left=dataAll, right=df,
                   left_index=True, right_index=True)
dataAll = dataAll.drop(columns=['Lyric_x'])
dataAll = dataAll.rename(columns={'SName':'Song', 'Lyric_y':'Lyric'})
```

```python
dataAll.tail(20)
dataAll.to_csv('data.csv')
```
